Create and label 4 separate charts for given equations.
You can use functions from the math module to help implement some of the equations above.

```python
import math

xsqrt = range(0,100)
ysqrt = [math.sqrt(x) for x in xsqrt]

plt.title('$\sqrt{x}$')
plt.plot(xsqrt,ysqrt)
```

```python
xcube = range(-100,100)
ycube = [x**3 for x in xcube]

plt.title('$x^3$')
plt.plot(xcube,ycube)
```

```python
xpow2 = range(-100,100)
ypow2 = [2**x for x in xpow2]

plt.title('$2^x$')
plt.plot(xpow2,ypow2)
```

```python
x = np.arange(-100,100,0.5)#range(-100,100)
y = [(1/(n+1)) if n != -1 else np.nan for n in x ]
plt.title('1/(x+1)')
plt.plot(x,y)
```

3. 
Combine the figures you created in the last step into one large figure with 4 subplots.

```python

plt.figure(figsize = (12,8))
# plot the first subplot
plt.subplot(221)
plt.plot(xsqrt, ysqrt)
plt.title('Sqrt x')

# the second subplot
plt.subplot(222)
plt.plot(xcube, ycube)
plt.title('x ^ 3')

# plot the third subplot
plt.subplot(223)
plt.plot(xpow2, ypow2)
plt.title('2^x')

# the fourth subplot
plt.subplot(224)
plt.plot(x, y)
plt.title('1/(x+1)')

plt.show()
```

#### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = np.arange(-10,10,0.5)

ysqrt = [math.sqrt(n) if n >= 0 else np.nan for n in x]

ycube = [n**3 for n in x]

ypow2 = [2**n for n in x]

y = [(1/(n+1)) if n != -1 else np.nan for n in x ]

plt.plot(x, ysqrt, label = '$\sqrt{x}$' )

plt.plot(x, ycube, label = '$x^3$')

plt.plot(x, ypow2, label = '$2^x$')

plt.plot(x, y, label = '1/(x+1)')
plt.title('Combined Plots')
plt.legend()
plt.xlim(-10,10)
plt.ylim(-10,10)


plt.show()
```

## Big O Notation 

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = range(0,100)

# Define each function using list comprehensions
O1 = [1 for n in x]
Ologn = [math.log(n) if n>0 else np.nan for n in x]
On = x
Onlogn = [n * math.log(n) if n>0 else np.nan for n in x]
On2 = [n**2 for n in x]
O2n = [2**n for n in x]
Onfact = [math.factorial(int(n))for n in x] # converts not integer values to integers - results in step plot
Onn = [n**n for n in x]

plt.rcParams['figure.figsize'] = [10, 7]

plt.xlim(0,10)
plt.ylim(0,100)

# plt.figure(figsize = (12,8)) Enabling this caused weird conflicts with setting axis limits. Better to set size using the rcParams above
plt.title('Big O Notation (linear scale)')
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.plot(x,O1, label = 'O(1)')
plt.plot(x,Ologn, label = 'O($\log{n}$)')
plt.plot(x,On, label = 'O(n)')
plt.plot(x,Onlogn, label = 'O(n $\log{n}$)')
plt.plot(x,On2, label = 'O(${n^2}$)')
plt.plot(x,O2n, label = 'O(${2^n}$)')
plt.plot(x,Onfact, label = 'O(${n!}$)')
plt.plot(x, Onn, label = 'O(${n^n}$)')
plt.legend(loc='upper right')
```

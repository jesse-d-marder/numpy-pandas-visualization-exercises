```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
x = range(-10,10)
O1 = [n+1 for n in x]
Ologn = [math.log(n) if n>0 else np.nan for n in x]
On = x
Onlogn = [n * math.log(n) if n>0 else np.nan for n in x]
On2 = [n**2 for n in x]
O2n = [2**n for n in x]
Onfact = [math.factorial(n) if n>0 else np.nan for n in x]
Onn = [n**n for n in x]


plt.ylim = (-10,10)

plt.plot(x,O1, label = 'O(1)')
plt.plot(x,Ologn, label = 'O($\log{n}$)')
plt.plot(x,On, label = 'O(n)')
plt.plot(x,Onlogn, label = 'O(n $\log{n}$)')
plt.plot(x,On2, label = 'O(${n^2}$)')
plt.plot(x,O2n, label = 'O(${2^n}$)')
plt.plot(x,Onfact, label = 'O(${n!}$)')
plt.plot(x, Onn, label = 'O(${n^n}$)')
plt.legend()

plt.show()
```
